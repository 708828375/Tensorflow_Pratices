```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# 加载电影评论数据
data = keras.datasets.imdb
```

```python
max_word = 10000
```

```python
# 进行数据加载
(x_train,y_train),(x_test,y_test) = data.load_data(num_words=max_word) # num_words指定编码单词的个数，此处表示只给前10000个单词进行编码
```

```python
x_train.shape,y_train.shape
```

```python
x_train[0] # 用一个整数来代替单词
```

将文本训练成为密集向量

```python
[len(x) for x in x_train]
```

```python
# 将数据长度填充到指定长度
x_train = keras.preprocessing.sequence.pad_sequences(x_train,300)
x_test = keras.preprocessing.sequence.pad_sequences(x_test,300)
```

```python
[len(x) for x in x_train]
```

```python
y_train
```

```python
model = keras.models.Sequential()
# 将整数映射成密集向量
# 参数一：输入数据的维度
# 参数二：映射向量的长度
model.add(layers.Embedding(10000,50,input_length=300))

#(25000,300,50)
# 给数据降维
# model.add(layers.Flatten())
model.add(layers.GlobalAveragePooling1D()) # 比Flatten更好的降维方法，实质是缩小模型的规模

model.add(layers.Dense(128,activation='relu',kernel_regularizer='l2'))
#model.add(layers.Dropout(0.5))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=0.001),
              loss='binary_crossentropy',
              metrics=['acc'])
```

```python
history = model.fit(x_train,y_train,epochs=10,batch_size=256,validation_data=(x_test,y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(history.epoch,history.history.get('acc'),'r',label='acc')
plt.plot(history.epoch,history.history.get('val_acc'),'b--',label='val_acc')
plt.legend()
```

```python
plt.plot(history.epoch,history.history.get('loss'),'r',label='loss')
plt.plot(history.epoch,history.history.get('val_loss'),'b--',label='val_loss')
plt.legend()
```

解决过拟合问题：1.Dropout   2.正则化L1,L2   3.缩小模型规模
